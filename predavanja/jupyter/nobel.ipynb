# Kako iz programa v Pythonu do baze

Delali bomo z bazo Nobelovih nagrajencev.

```python
import sqlite3 as dbapi

BAZA = 'nobel.db'
```

## Izvajanje skripte SQL

Če imamo pripravljeno skripto SQL, jo lahko preberemo iz Pythona in izvedemo z metodo `executescript`.

```python
def uvozi_skripto(skripta, baza):
    try:
        conn = dbapi.connect(baza)
        cur = conn.cursor()

        with open(skripta) as f:
            with conn:
                cur.executescript(f.read())
    finally:
        conn.close()
```

```python
uvozi_skripto('nobel.sql', BAZA)
```

Pripravimo še funkcijo, ki bo v tabelo `nobel` vpisala vrstico s podanimi podatki.

```python
def vpisi_nagrajenca(cur, **kwargs): # uporaba poimenovanih parametrov
    '''Vpiši osebo z danimi podatki.'''
    cur.execute("""
          INSERT INTO nobel (yr, subject, winner)
          VALUES (:yr, :subject, :winner)
        """, kwargs)
    return cur.lastrowid # vrnemo ID vpisane vrstice
```

Sedaj lahko zapišemo funkcijo za ustvarjanje baze in uvoz podatkov, ter jo nato izvedemo.

```python
def uvozi_nobel():
    try:
        conn = dbapi.connect(BAZA)
        cur = conn.cursor()

        with conn:
            cur.execute(CREATE_TABLE_nobel)
            for leto, podrocje, dobitnik in nagrajenci:
                id = vpisi_nagrajenca(cur, yr=leto, subject=podrocje, winner=dobitnik)
                print(f"Vrstica {id}: dobitnik za področje {podrocje} leta {leto} je {dobitnik}.")
    finally:
        conn.close()
```

## Uporaba baze

Napišimo še funkcijo, ki nam izpiše Nobelove nagrajence za določeno leto.

```python
def nobel_leto(leto):
    try:
        conn = dbapi.connect(BAZA)
        cur = conn.cursor()
        cur.execute("""
              SELECT subject, winner FROM nobel
               WHERE yr = ?
            """, [leto])
        for podrocje, dobitnik in cur:
            print(f"Dobitnik za področje {podrocje} je {dobitnik}.")
    finally:
        conn.close()
```
